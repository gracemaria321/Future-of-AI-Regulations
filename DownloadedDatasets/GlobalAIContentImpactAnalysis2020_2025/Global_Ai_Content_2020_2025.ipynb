## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the Dataset

```python
path="/content/Global_AI_Content_Impact_Dataset (1).csv"

path
```

```python
df=pd.read_csv(path)
df
```

### Dataset Structure and Info

```python
df.info()
```

## Null Value Check

```python
df.isnull().sum()
```

### Histrogram

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['AI Adoption Rate (%)'], kde=True, bins=20, color='blue')
plt.title("Distribution of AI Adoption Rate (%)")
plt.xlabel("AI Adoption Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

### Barplot

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Consumer Trust in AI (%)', data=df, color='red')
plt.xticks(rotation=90)  #
plt.title('Consumer Trust in AI (%) by Country')
plt.xlabel('Country')
plt.ylabel('Consumer Trust in AI (%)')
plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()
```

### boxplot

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Industry',y='AI-Generated Content Volume (TBs per year)')
plt.xticks(rotation=90)
plt.title('AI-Generated Content Volume By Industry')
plt.tight_layout()
plt.show()
```

### Lineplot

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=df, x='Country', y='Revenue Increase Due to AI (%)', hue='Year', palette='pastel',marker='o')
plt.xticks(rotation=60)
plt.xlabel('Country')
plt.ylabel('Revenue Increase Due to AI (%)')
plt.title('Revenue Increase Using AI By Country')
plt.show()
```

### Scatterplot

```python
plt.figure(figsize=(16,5))
plt.scatter(df['AI Adoption Rate (%)'], df['Revenue Increase Due to AI (%)'])
plt.xlabel('AI Adoption Rate (%)')
plt.ylabel('Revenue Increase Due to AI (%)')
plt.title('AI Adoption Rate vs Revenue Increase')
plt.grid(True)
plt.show()
```

```python
industry_job_loss = df.groupby('Industry')['Job Loss Due to AI (%)'].mean()
industry_job_loss
```

## pie chart

```python
plt.figure(figsize=(12,6))
plt.pie(industry_job_loss, labels=industry_job_loss.index,autopct='%1.1f%%')
plt.title('Average Job Loss Due to AI by Industry')
plt.show()
```

## Countplot

```python
plt.figure(figsize=(10,6))
sns.countplot(y=df['Top AI Tools Used'], order=df['Top AI Tools Used'].value_counts().index, palette='viridis')
plt.xlabel('Count')
plt.ylabel('AI Tools')
plt.title('Most Popular AI Tools Across Industries')
plt.show()
```

## Heatmap

```python
plt.figure(figsize=(14,6))
correlation_matrix=df.corr(numeric_only=True)
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heatmap of AI Impact Features')
plt.show()
```
